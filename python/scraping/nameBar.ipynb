```python
fileName = r'/home/liang.liang/code/Tools/python/scraping/tianjin_full.xlsx'
#fileName = r'/Users/lc/code/Tools/python/scraping/tianjin_full.xlsx'

data = xlrd.open_workbook(fileName)

table = data.sheets()[0]


Source =  {"name":table.col_values(1,1)}

Total = len(Source["name"])

nameList = {}
```

```python
Total
```

```python
for i in range(Total):
    name = Source["name"][i]
    if(name in nameList):
        if(name == "流拍"):
            pass
        elif(name == "无名氏"):
            pass
        else:
            nameList[name] +=1
    else:
        nameList[name] = 1
```

```python
keys = list(nameList.keys())
values = list(nameList.values())
```

```python
newNameList = {}
```

```python
for name in keys:
    if ( nameList[name] > 1 ):
        newNameList[name] = nameList[name]

        
        
```

```python
newKeys = list(newNameList.keys())
newValues = list(newNameList.values())
print(newKeys)
print(newValues)

```

```python
number = len(newKeys)
print(number)
```

```python
bar = Bar()

bar.add_xaxis(newKeys)
bar.add_yaxis("buyer",newValues,category_gap="50%")
bar.reversal_axis()
bar.set_series_opts(label_opts=opts.LabelOpts(is_show=False))

bar.set_global_opts(
    title_opts=opts.TitleOpts(title="123"),
    #xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-90)),
    #datazoom_opts=opts.DataZoomOpts()
    yaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=0),
                             name = "",
                             is_inverse = False,
                             name_location = "end",
                             name_gap = 5,
                             name_rotate = 15,
                             interval = 30,  # 无法在类目轴中使用
                             split_number = 180,
                             min_interval = 1
                            )
)

bar.render_notebook()
```
