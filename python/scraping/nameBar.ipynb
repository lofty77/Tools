```python
from pyecharts.engine import create_default_environment
from pyecharts import Page
from pyecharts import Bar, Line, Grid
from pyecharts import Liquid

import xlrd
```

```python
fileName = r'/home/liang.liang/code/Tools/python/scraping/name2018_2019.5.xlsx'

data = xlrd.open_workbook(fileName)

table = data.sheets()[0]


Source =  {"name":table.col_values(1,1)}

Total = len(Source["name"])

nameList = {}
```

```python
for i in range(Total):
    name = Source["name"][i]
    if(name in nameList):
        nameList[name] +=1
    else:
        nameList[name] = 1
```

```python
keys = list(nameList.keys())
values = list(nameList.values())
```

```python
newNameList = {}
```

```python
for name in keys:
    if ( nameList[name] > 1):
        newNameList[name] = nameList[name]
        
        
```

```python
newKeys = list(newNameList.keys())
newValues = list(newNameList.values())
print(newKeys)
print(newValues)

```

```python
number = len(newKeys)
print(number)
```

```python
bar = Bar(title="Name",subtitle ="2018.1_2019.5",width = number*10, height = number*30 )
#bar = Bar(title="Name ",subtitle ="2018.1_2019.5")
```

```python
bar.add(0, newKeys, newValues
            ,xaxis_rotate=0,yaxis_rotate=0,xaxis_interval= 0
            ,is_label_show=False,yaxis_pos = "left"
            ,xaxis_type = 'category',is_stack = True,is_convert = True,yaxis_label_textsize = 12
            ,yaxis_name_gap = 5
            ,is_toolbox_show = True,is_more_utils=True
            ,yaxis_interval = 0,yaxis_margin= 0
            ,legend_pos="30%"
           )
```

```python
bar.render("namelist.html")
```
